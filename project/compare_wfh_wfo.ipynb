```python
## compare the average salaries for jobs that offer both wfh and wfo that also have more than 2 job posting

import pandas as pd
import matplotlib.pyplot as plt
import ast

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

file = "C:\\Users\\aksha\\data_analysis_job_market\\dataset\\data_jobs.csv"
df = pd.read_csv(file)

job_counts = df['job_title_short'].value_counts()

valid_titles = job_counts[job_counts > 2].index

df= df[df['job_title_short'].isin(valid_titles)]

df = pd.pivot_table(df,
                    index= 'job_title_short',
                    columns = 'job_work_from_home',
                    values = 'salary_year_avg',
                    aggfunc='mean')


df.head()
```

```python
## lets plot thse to find how the salaries compare

df.plot(kind='bar')
plt.title('salaries for remote vs non remote jobs')
```
