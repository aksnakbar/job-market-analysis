```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)


# import dataset from a csv file and create a dataframe from it

df = pd.read_csv('C:\\Users\\aksha\\data_analysis_job_market\\dataset\\data_jobs.csv')

# data needs to be cleaned to have the right datatype inorder to perform relevant analysis
# convert the date of job postings into a date time object to be able to use the vaious date time attribute to perform further analysis

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['Month'] = df['job_posted_date'].dt.month
df['Month_name']= df['job_posted_date'].dt.month_name

# convert the skills in the skill column from a string into a list
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)

df.head(10)
```

```python
# let's create a subset of the above data frame to only have jobs that have been posted in the United States

df_US = df[df['job_country']=='United States'].copy()
df_US.head(10)
```

```python
# let's filter it down to US data analyst roles 

df_DA_US = df_US[df_US['job_title_short']=='Data Analyst'].copy()
df_DA_US.head(10)
```

```python
# let's find the different job locations posted using value_counts() which is grouping by location and counting the number 
# of jobs posted under each location.job_posted_per_location is a sries with index as job_location and count as the column

jobs_posted_per_location = df_DA_US['job_location'].value_counts()
jobs_posted_per_location.head(10)
```

```python
# In order to plot the above to visulaze the relationship between the location jobs were posted and the number of jobs posted/location
# I want to convert the above series into a dataframe


jobs_posted_per_location = jobs_posted_per_location.to_frame()
```

```python
df_plot = jobs_posted_per_location.head(10)
```

```python

sns.set_theme(style='ticks')
sns.barplot(data=df_plot , x='count', y='job_location', hue='count', palette = 'dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs by Title in U.S. Cities with the Highest Job Concentration')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Let's check what share of all the jobs :
# 1.Are work from home 
# 2.Doesn't need a degree 
# 3.Offers health insurance

df_wfh_summary = df_DA_US.groupby('job_work_from_home')['job_title'].count()
df_wfh_summary= df_wfh_summary.rename(index = {False:'Non-Remote',True:'Remote'},axis=0)
type(df_wfh_summary)
```

```python
df_wfh_summary.plot(kind='pie', autopct='%2.2f%%', 
                    title='Percentage of Data Analyst jobs that offer work from home',
                    startangle=90,
                    )
plt.xlabel('')
plt.ylabel('')
```

```python
df_degree_summary = df_DA_US.groupby('job_no_degree_mention')['job_title'].count()
df_degree_summary = df_degree_summary.rename({True:'Degree Required',False:'Degree not required'})
```

```python
df_health_insurance_summary = df_DA_US.groupby('job_health_insurance')['job_title'].count()
df_health_insurance_summary = df_health_insurance_summary.rename(index={True:'Insurance provide',False:'Insurance not provided'})
```

```python
# I want a sing;e figure with 3 pie charts 

fig,ax = plt.subplots(1, 3, figsize=(15,5))
custom_colors=['dodgerblue','hotpink']

ax[0].pie(df_wfh_summary, labels=df_wfh_summary.index, autopct='%2.2f%%',startangle=90,textprops={'fontsize': 12,'fontweight':'bold'},colors=custom_colors)
ax[0].set_title('Percentage of Data Analyst jobs offering Work from Home',fontsize=12,fontweight='bold')
ax[1].pie(df_degree_summary, labels=df_degree_summary.index, autopct='%2.2f%%',startangle=90,textprops={'fontsize': 12,'fontweight':'bold'},colors=custom_colors)
ax[1].set_title('Percentage of Data Analyst jobs needing a Degree',fontsize=12,fontweight='bold')
ax[2].pie(df_health_insurance_summary,labels=df_health_insurance_summary.index, autopct='%2.2f%%',startangle=90,textprops={'fontsize': 12,'fontweight':'bold'},colors=custom_colors)
ax[2].set_title('Percentage of Data Analyst jobs offering Health insurance',fontsize=12,fontweight='bold')

fig.tight_layout()
```

```python
# Let's find the number of Data Analyst jobs offered by each company

df_company = df_DA_US['company_name'].value_counts()
df_company = df_company.to_frame().head(10)
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_company , x='count', y='company_name', hue='count', palette = 'dark:r_r', legend=False)
sns.despine()
plt.suptitle('Number of Jobs by Company in U.S. Cities for Data Analyst jobs',fontsize=14)
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
plt.tight_layout(rect=[0, 0, 1, 0.95])
```
