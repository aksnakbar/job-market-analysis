```python
#import data from the csv file

import pandas as pd
import matplotlib.pyplot as plt

file = "C:\\Users\\aksha\\data_analysis_job_market\\dataset\\data_jobs.csv"
df = pd.read_csv(file)

#1.clean the data by converting the job_posted_date into a datetime object from a string

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df.job_posted_date.dt.month

#2.Clean by filling all the job postings with NaN for salary with the median salary on both yearly and hourly salary columns

median_salary_yearly = df['salary_year_avg'].median()
median_salary_hourly = df['salary_hour_avg'].median()

df_filled = df.copy()

df_filled['salary_hour_avg']=df_filled['salary_hour_avg'].fillna(median_salary_hourly)
df_filled['salary_year_avg']=df_filled['salary_year_avg'].fillna(median_salary_yearly)

df_filled.sample(10)


```

```python
# we want plot only for top six countries 

top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
                        values = 'salary_year_avg',
                        index = 'job_country',
                        columns = 'job_title_short',
                        aggfunc= 'median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst','Data Engineer','Data Scientist']
df_job_country_salary = df_job_country_salary[job_titles]
df_job_country_salary.plot(kind = 'bar')
plt.ylabel('Median salary ($USD)')
plt.xlabel("")
plt.title('Median salary by country and job titile')
plt.xticks(rotation=45, ha='right')
plt.show()
```
