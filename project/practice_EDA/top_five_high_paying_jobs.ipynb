```python

import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

## read the  date from a csv file into dataframe

file_name = "C:\\Users\\aksha\\data_analysis_job_market\\dataset\\data_jobs.csv"
df = pd.read_csv(file_name)
df.head()

## create another dataframe that only has US jobs with workfrom home 

df_US_wfh = df[(df['job_country']== 'United States') & (df['job_work_from_home']== True)]
df_US_wfh.head()
```

```python

number_of_jobpostings = df_US_wfh['job_title'].value_counts()

skill_average_salary = df_US_wfh.groupby(['job_title'])['salary_year_avg'].mean()



df_summary =pd.DataFrame({
    'job_postings':number_of_jobpostings,
    'average_salary_per_skill':skill_average_salary
})

highest_salaries = df_summary.sort_values(by='job_postings',ascending = False)
highest_salaries.head(5)
```
