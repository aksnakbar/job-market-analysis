```python
## find the top 5 paying jobs in the US that are remote

import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

## read the  date from a csv file into dataframe

file_name = "C:\\Users\\aksha\\data_analysis_job_market\\dataset\\data_jobs.csv"
df = pd.read_csv(file_name)
df.head()

## create another dataframe that only has US jobs with workfrom home 

df_US_wfh = df[(df['job_country']== 'United States') & (df['job_work_from_home']== True)]
df_US_wfh.head()


## highest paying salary so I have to sort by value in descending order

df_sorted_values= df_US_wfh.sort_values('salary_year_avg',ascending=False)
df_sorted_values.head()


```
